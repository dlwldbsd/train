```python
#네이버에 검색결과 페이지 가져오기
import requests
a = '야구' #검색어
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={}' #검색어부분 지우고 빈 중괄호 넣어주기
r = requests.get(url.format(a)) #중괄호에 a넣기
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser') #파씽 = 한 부모씩 잘라줘
```

```python
prd_link = soup.find('a', attrs={'class':'match'})
print(prd_link)
```

```python
prd_link['href']
```

href? 상세페이지 이동 데이터

```python
url ='https://search.naver.com/search.naver'+prd_link['href']
r=requests.get(url)
```

```python
print(r.text)
```
